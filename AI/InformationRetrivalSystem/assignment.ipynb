```python
from bs4 import BeautifulSoup
import requests

url='https://en.wikipedia.org/wiki/Wikipedia:Featured_articles'
res = requests.get(url)
soup =BeautifulSoup(res.text , 'html.parser')
links=soup.find_all('a')[83:1083]
links1 = []

for l in links:
    link=l.get('href')
    link=str(link)
    links1.append("https://en.wikipedia.org"+link)
links = links1
```

```python
for i in range(0, len(links)):
    datawrite=requests.get(links[i])
    soup=BeautifulSoup(datawrite.text,'html.parser')
    file=open('file'+str(i)+'.txt', 'w', encoding=('utf-8'))
    heading = soup.find('h1')
    file.write(heading.text+'\n')
    for para in soup.find_all('p')[0:2]:
        file.write(para.text+'\n')
    file.close()
```

```python
import numpy as np

freq=dict()
wtn=dict()
ntw=dict()

for i in range(0, 1000):
    file=open('file'+str(i)+'.txt', 'r', encoding=('utf-8'))
    words = file.read().replace(',' or ';' or ':'or']' or '[' or '}' or '{' or ')' or '('or '"' or '?' or '!'or'@' or '#', '')
    for word in words.split():
        word = word.lower()
        if word not in freq:
            freq[word] = 1
        else:
            freq[word] = freq[word] + 1
    file.close()

size_of_vocabulary = len(freq
                        )
keys=freq.keys()
i=0

for k in keys:
    wtn[k]=i
    ntw[i]=k
    i=i+1 
```

```python
document=np.zeros((1000,size_of_vocabulary),dtype=int)
for r in range(0, 1000):
    file=open('file'+str(r)+'.txt', 'r', encoding=('utf-8'))
    words = file.read().replace(',' or ';' or ':'or']' or '[' or '}' or '{' or ')' or '('or '"' or '?' or '!'or'@' or '#', '')
    for word in words.split():
        word = word.lower()
        c = wtn[word]
        document[r,c]=1
```

```python
query=input("Search something: ")

vector=np.zeros((size_of_vocabulary),dtype=int)

query=query.replace(',' or ';' or ':'or']' or '[' or '}' or '{' or ')' or '('or '"' or '?' or '!'or'@' or '#', '')
for word in query.split():
    word=word.lower()
    if word in wtn:
        c=wtn[word]
        vector[c]=1

rank_matrix=vector.dot(document.T)

sorted_rank_matrix=np.sort(rank_matrix)
reverse_sorted_rank_matrix=sorted_rank_matrix[::-1]

sorted_file_matrix=np.argsort(rank_matrix)
reverse_sorted_file_matrix=sorted_file_matrix[::-1]

print("Result by ranking:")
for i in reverse_sorted_file_matrix:
    print('file'+str(i)+'.txt')
```
