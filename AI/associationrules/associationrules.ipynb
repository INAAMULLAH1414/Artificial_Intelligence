```python
import pandas as pd

def loadDataSet():
    data=pd.read_csv('data.csv',header=None)
    return (data.values.tolist())
print(loadDataSet())
def minSupport():
    ms=float(input('Enter minimum support:'))
    return ms
def minconfidence():
    mc=float(input('Enter minimum confidence:'))
    return mc
def createC1(dataSet):
    C1 = []
    for t in dataSet:
        for item in t:
            if not [item] in C1:
                C1.append([item])
                
   
    return list(map(frozenset, C1))
def scanData(D, Ck, minSupport):
    a = {}
    for tid in D:
        for can in Ck:
            if can.issubset(tid):
                if not can in a: a[can]=1
                else: a[can] += 1
    numItems = float(len(D))
    L= []
    supportData = {}
    for key in a:
        support = a[key]/numItems
        if support >= minSupport:
            L.insert(0,key)
        supportData[key] = support
    return L, supportData
dataSet = loadDataSet()
#print(dataSet)
C1 = createC1(dataSet)


D = list(map(set,dataSet))
def createck(Lk, k): 
    L = []
    lengthLk = len(Lk)
    for i in range(lengthLk):
        for j in range(i+1, lengthLk): 
            L1 = list(Lk[i])[:k-2]; L2 = list(Lk[j])[:k-2]
            L1.sort(); L2.sort()
            if L1==L2:                  #if first k-2 elements are equal
                L.append(Lk[i] | Lk[j])
    return L
def apriori(dataSet, minSupport):
    C1 = createC1(dataSet)
    D = list(map(set, dataSet))
    L1, supportData = scanData(D, C1, minSupport)
    L = [L1]
    k = 2
    while (len(L[k-2]) > 0):
        Ck = apriori(L[k-2], k)
        Lk, supK = scanData(D, Ck, minSupport)#scan and generate Lk
        supportData.update(supK)
        L.append(Lk)
        k += 1
    return L, supportData
a=minSupport()
L,suppData = apriori(dataSet,a)
def generateRules(L, supportData, minConfidence):  #supportData is a dict coming from scanD
    b = []
    for i in range(1, len(L)):
        for freqSet in L[i]:
            H1 = [frozenset([item]) for item in freqSet]
            if (i > 1):
                rulesFromConseq(freqSet, H1, supportData, b, minConfidence)
            else:
                calculateConfidence(freqSet, H1, supportData,b, minConfidence)
    return b        
def calculateConfidence(freqSet, H, supportData, brl, minConfidence):
    d = [] #create new list to return
    for conseq in H:
        conf = supportData[freqSet]/supportData[freqSet-conseq] 
        if conf >= minConfidence: 
            print (freqSet-conseq,'-->',conseq)
            brl.append((freqSet-conseq, conseq, conf))
            d.append(conseq)
    return d
def rulesFromConseq(freqSet, H, supportData, brl, minConfidence):
    m = len(H[0])
    if (len(freqSet) > (m + 1)): 
        p = createck(H, m+1)#create new candidates
        p = calcConfidence(freqSet, p, supportData, brl, minConfidence)
        if (len(p) > 1):    
            rulesFromConseq(freqSet, p, supportData, brl, minConfidence)
L,suppData= apriori(dataSet,a)
rules= generateRules(L,suppData, minconfidence())
```
